```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("datasets/suicide-blasts-dataset-cleaned.csv")
```

```python
df['Total Killed'] = df['Killed Max']
df['Total Injured'] = df['Injured Max']
df['Total Casualties'] = df['Total Killed'] + df['Total Injured']
```

```python
# 1. Number of incidents per year
incidents_per_year = df.groupby('Year').size()
plt.figure(figsize=(10,5))
plt.plot(incidents_per_year.index, incidents_per_year.values, marker='o')
plt.title("Number of Incidents per Year")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.grid(True)
plt.show()
```

```python
# 2. Top cities with most incidents
top_cities = df['City'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_cities.plot(kind='bar', color='skyblue')
plt.title("Top 10 Cities with Most Incidents")
plt.xlabel("City")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Casualties by year
casualties_per_year = df.groupby('Year')['Total Casualties'].sum()
plt.figure(figsize=(10,5))
plt.bar(casualties_per_year.index, casualties_per_year.values, color='salmon')
plt.title("Total Casualties by Year")
plt.xlabel("Year")
plt.ylabel("Total Casualties")
plt.show()
```

```python
# 4. Target type distribution
target_counts = df['Target Type'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Target Types")
plt.show()
```

```python
# 5. Scatter plot of casualties vs. suicide blasts
plt.figure(figsize=(10,6))
plt.scatter(df['No. of Suicide Blasts'], df['Total Casualties'], alpha=0.6, c='purple')
plt.title("Casualties vs. Number of Suicide Blasts")
plt.xlabel("No. of Suicide Blasts")
plt.ylabel("Total Casualties")
plt.grid(True)
plt.show()
```
