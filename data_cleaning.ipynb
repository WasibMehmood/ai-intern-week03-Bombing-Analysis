```python
import pandas as pd
data = pd.read_csv('datasets/suicide-blasts-dataset.csv')
```

```python
data.head(10)
```

# PREPROCESSING OF DATA

```python
data['Date'] = pd.to_datetime(data['Date'] ,errors = 'coerce')
```

```python
data['Killed Min'] = pd.to_numeric(data['Killed Min'],errors = 'coerce').fillna(0)
data['Killed Max'] = pd.to_numeric(data['Killed Max'],errors = 'coerce').fillna(0)
data['Injured Min'] = pd.to_numeric(data['Injured Min'],errors = 'coerce').fillna(0)
data['Injured Max'] = pd.to_numeric(data['Injured Max'],errors = 'coerce').fillna(0)
```

```python
missing = data.isna().sum()
missing
```

```python

for col in ['Latitude', 'Location','Hospital Names']:
    if col in data.columns:
        data[col] = data[col].astype(str).str.strip()
for col in ['Influencing Event/Event', 'Longitude']:
    if col in data.columns:
        data[col] = data[col].astype(str).str.strip()
```

```python
for col in data.select_dtypes(include='object').columns:
    data[col].fillna(data[col].mode()[0], inplace=True)
    
for col in data.select_dtypes(include='number').columns:
    data[col].fillna(0, inplace=True)
int_cols = ['Killed Min', 'Killed Max', 'Injured Min', 'Injured Max', 'No. of Suicide Blasts']
for col in int_cols:
    if col in data.columns:
        data[col] = data[col].astype(int)
```

```python
columns_to_drop = ['Temperature(F)', 'Temperature(C)','Explosive Weight (max)','Islamic Date','Time']
data.drop(columns=[col for col in columns_to_drop if col in data.columns], inplace=True)
```

```python
missing = data.isna().sum()
missing
```

```python
data.to_csv('suicide-blasts-dataset-cleaned.csv', index = False)
```
